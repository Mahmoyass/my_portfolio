```python
sns.countplot(x="Branch",data=df,palette="pastel",hue="Product line")
plt.title("count of sales of each branch ")
plt.show()
```

```python
sns.countplot(x="Customer type",data=df,palette="pastel",hue="Branch")
plt.title("Customer type by Branch")
plt.show()
```

```python
sns.countplot(x="Gender",data=df,palette="pastel",hue="Branch")
plt.title("count of gender by Branch ")
plt.show()
```

```python
sns.countplot(x="Payment",data=df,palette="pastel",hue="Branch")
plt.title("methods of payment of each branch ")
plt.show()
```

```python
sns.boxplot(x="Branch",y="gross income",data=df)
plt.title("gross income of each branch ")
plt.show()
```

```python
df["City"].value_counts().plot.pie()
plt.title("  Cities of the branches ")
plt.show()
```

```python
sns.kdeplot(data=df,x="Rating",multiple="stack",hue="Branch")
plt.title("Rating of the Branches ")
plt.show()
```

```python
sns.boxplot(x="Product line",y="Unit price",data=df)
plt.title("Unit price of each Product ")
plt.show()
```

```python
sns.displot(df["gross income"],kde=True,bins=7)
plt.title("Distribution of gross income  ")
plt.show()
```

```python
sns.displot(df["Unit price"],kde=True,bins=7)
plt.title("Distribution of Unit price  ")
plt.show()
```
