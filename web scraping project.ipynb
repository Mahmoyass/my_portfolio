# getting url

```python
url="https://wuzzuf.net/search/jobs/?q=illustrator&a=navbl"
```

# connection

```python
client= urlopen(url)
```

# getting the html code to the full page

```python
html= client.read()
html
```

# cloing the request

```python
client.close()
```

# creating an html parser using beatifulsoup

```python
soup=bs(html,"html.parser")
soup
```

# creating container for needed data

```python
containers= soup.find_all("div",{"class":"css-1gatmva e1v1l3u10"})
```
