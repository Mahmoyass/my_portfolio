```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
sns.set_palette("rainbow")
sns.set_style("whitegrid")
```

```python
df=pd.read_csv("Tomato.csv")
df
```

```python
df["Date"]=pd.to_datetime(df["Date"])
```

```python
df.info()
```

```python
df
```

```python
df.set_index(["Date"],inplace=True)
```

```python
df
```

```python
df.plot(figsize=(20,10))
plt.title("correlation between Minimum,Maximum and Average over the time ")
plt.show()
```

```python
df["Minimum"].plot(figsize=(20,10))
plt.title("Minimum prices over the time ")
plt.show()
```

```python
df["Maximum"].plot(figsize=(20,10))
plt.title("Maximum prices over the time ")
plt.show()
```

```python
df["Average"].plot(figsize=(20,10))
plt.title("Average prices over the time ")
plt.show()
```

```python
df[["Average"]].rolling(12).mean().plot(figsize=(20,10))
plt.title("Average prices over the time ")
plt.show()
```
